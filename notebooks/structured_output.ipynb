```python
from pydantic import BaseModel,Field
from typing import Optional,Literal
from langchain_groq import ChatGroq


## Literal use to expect two values true or false.
```

```python
review = "I recently bought a noise cancelling headphone and I’m really impressed with the performance. The sound quality is crystal clear, and the noise cancellation works surprisingly well even in crowded environments. Battery life is also decent, lasting almost a full day on a single charge. Overall, it feels like a great value for money and I would definitely recommend it to others. Rating: 5/5."
```

```python
class Review(BaseModel):
    Product:str = Field(description="Name of the product being reviewed")
    Pros: list[str] = Field(description="List of pros about the product")
    cons: list[str] = Field(description="List of cons about the product")
    sentiment: Literal["positive", "negative"] = Field(description="Overall sentiment of the review")
    rating: float = Field(ge=1, le=5, description="Rating given by the reviewer out of 5")
    reviewer: Optional[str] = Field(description="Name of the reviewer, if available",default=None)
```

```python
model = ChatGroq(model="llama-3.1-8b-instant",temperature=0.2)
```

```python
structModel = model.with_structured_output(Review)
```

```python
res = structModel.invoke(review)
```
