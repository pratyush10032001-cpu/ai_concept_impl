```python
from langchain_core.prompts import ChatPromptTemplate,MessagesPlaceholder
from langchain_groq import ChatGroq
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_community.chat_message_histories import ChatMessageHistory
from dotenv import load_dotenv
```

```python
llm = ChatGroq(model="llama-3.1-8b-instant")
```

```python
prompt = ChatPromptTemplate.from_messages([('system','You are an Friendly ChatBot you have access to the past conversion history.'),MessagesPlaceholder(variable_name='history'),('human','{input}')])
```

```python
parser = StrOutputParser()
```

```python
store = {}
```

```python
chain = prompt | llm | parser
```

```python
def get_history(session_id:str):
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]  
```

```python
chain_with_history = RunnableWithMessageHistory(chain,get_history,input_messages_key='input',history_messages_key='history',output_messages_key='output')
```

```python
res1 = chain_with_history.invoke({'input':"Hello I am Pratyush"},config={'configurable':{'session_id':"pratyush1234"}})
```

```python
print('Bot',res1)
```
