```python
from langchain_core.prompts import PromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_groq import ChatGroq
from dotenv import load_dotenv
from langchain_core.output_parsers import StrOutputParser
```

```python
## Simple Sequential Chain
```

```python
prompt = PromptTemplate(template='Create a 5 point summary on {Topic}',input_variables=['Topic'])
```

```python
parser = StrOutputParser()
```

```python
model = ChatGoogleGenerativeAI(model='gemini-2.5-flash')
```

```python
## chain = prompt model parser connects ----------> it known as LCEL (langchain expression language)

chain = prompt | model | parser 
```

```python
result = chain.invoke({'Topic':'Cricket Asia Cup '})
```
