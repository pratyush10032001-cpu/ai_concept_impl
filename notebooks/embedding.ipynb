### Implementation of embedding with huggingface and check the query consine similarity with the embbed documents. 

```python
from langchain_huggingface import HuggingFaceEmbeddings
from sklearn.metrics.pairwise import cosine_similarity as consine_similarity
```

```python
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
text = ["csk with 10 wins in 14 matches is the best team in the IPL history","rcb is loser team in the IPL history with 4 wins in 14 matches","mi is second best team in the IPL history with 9 wins in 14 matches"]
```

```python
embeddings_list = embeddings.embed_documents(text)
```

```python
embeddings_list
```

```python
query = "which team is best in the IPL history?"
query_embedding = embeddings.embed_query(query)
```

```python
result_similarity = consine_similarity([query_embedding], embeddings_list)
```

```python
result_similarity[0]
```

```python
index,value = sorted(list(enumerate(result_similarity[0])), key=lambda x: x[1])[-1]

print(f"the most similar sentence is {text[index]} with similarity score of {value}")
```

### working with chroma DB

```python
from langchain.vectorstores import Chroma
from langchain.schema import Document
```

```python
docs=[
    Document(page_content="srk is famous bollywood actor,he has 100 movies,won multiple awards", metadata={"actor":"sharukh khan","source": "bollywood"}),
    Document(page_content="rajnikath is a famous tamil actor,he has 50 movies,won multiple awards", metadata={"actor": "rajnikanth", "source": "tamil cinema"})
]
```

```python
vector_store = Chroma(embedding_function=HuggingFaceEmbeddings(),persist_directory="chroma_db",collection_name="actors")
```

```python
## insertion of documents into chroma vector database

# add documents array into chroma vector database
vector_store.add_documents(docs)
```

```python
## reading from chroma vector database

vector_store.get(include=['documents','embeddings','metadatas'])
```

```python
vector_store.similarity_search(query="which actor is from bollywood ?", k=2)
```

```python
updated_srk_doc = Document(page_content="srk is famous bollywood actor,he has 100 movies,won multiple awards.He is also known as king khan", metadata={"actor":"sharukh khan","source": "bollywood"})
```

```python
### update the chroma vector database with new document content for srk

vector_store.update_document(document_id="2a9fccca-faee-428a-9b0e-38ef8316ce81", document=updated_srk_doc)
```

```python
vector_store.get(include=['documents','embeddings','metadatas'])
```

```python
## delete the document for srk from chroma vector database

vector_store.delete(ids=["2a9fccca-faee-428a-9b0e-38ef8316ce81"])
```
