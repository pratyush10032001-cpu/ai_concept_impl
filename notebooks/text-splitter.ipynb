### Text splitter 

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter,Language
```

```python
### length based splitting implementation

text_splitter = RecursiveCharacterTextSplitter(chunk_size=100,chunk_overlap=20,separators=" ")

text = "This is a sample text to demonstrate the length based splitting implementation. The text will be split into chunks of specified size with some overlap between the chunks."

chunk_text = text_splitter.split_text(text)

chunk_text[1]
```

```python
## text structure based splitting implementation

text_splitter = RecursiveCharacterTextSplitter(chunk_size=100,chunk_overlap=0,separators=" ")

text = "This is a sample text to demonstrate the length based splitting implementation. The text will be split into chunks of specified size with some overlap between the chunks."

chunk_text = text_splitter.split_text(text)

len(chunk_text[0])

```
