```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_groq import ChatGroq
from dotenv import load_dotenv
import google.genai as genai
import os
```

```python
load_dotenv()
```

```python
google_model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
groq_model = ChatGroq(model="llama-3.1-8b-instant",temperature=0.2) 
```

```python
prompt1 = PromptTemplate(template="Generate a Detailed Report on the {topic}?",input_variables=["topic"])
```

```python
prompt2 = PromptTemplate(template="Create a five point Summary of this {Report}?",input_variables=["Report"])
```

```python
final1 = prompt1.invoke({'topic':"Google vs Groq"})
```

```python
## chain is used to create pipeline for llms 

parser = StrOutputParser()

chain = prompt1 | google_model | parser | prompt2 | groq_model | parser
```

```python
response = chain.invoke({'topic':"Google vs Groq"})
```
