```python
from langchain_core.runnables import RunnablePassthrough,RunnableParallel
from langchain_core.prompts import PromptTemplate
from langchain_groq import ChatGroq
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
prompt = PromptTemplate(template="Generate a joke on this {Topic}",input_variables=['Topic'])
```

```python
llm = ChatGroq(model="llama-3.1-8b-instant")
```

```python
parser = StrOutputParser()
```

```python
chain1 = RunnableSequence(prompt,llm,parser)
```

```python
prompt2 = PromptTemplate(template='Generate the Explanation for the following {joke}',input_variables=['joke'])
```

```python
explain = prompt | llm | parser | prompt2 | llm | parser
```

```python
explain.invoke({'Topic':'AI vs Humans'})
```

```python
parallel = RunnableParallel({'joke':RunnablePassthrough(),'Explanation':prompt2 | llm | parser})
```

```python
final_chain = chain1 | parallel
```

```python
final_chain.invoke({'Topic':'AI vs Humas'})
```

```python
from langchain_core.runnables import RunnablePassthrough,RunnableLambda
from langchain_core.prompts import PromptTemplate
from langchain_groq import ChatGroq
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv
```

```python
prompt = PromptTemplate(template="Generate a joke on this {Topic}",input_variables=['Topic'])
```

```python
def wordN(text):
    return len(text.split())
```

```python
llm = ChatGroq(model="llama-3.1-8b-instant")
```

```python
chainJ = prompt | llm | parser
```

```python
parser = StrOutputParser()
```

```python
parallelRunnables = RunnableParallel({'joke':RunnablePassthrough(),'words count':RunnableLambda(wordN)})
```

```python
final_chain = chainJ | parallelRunnables
```

```python
final_chain.invoke({'Topic':'AI vs Humans'})
```
